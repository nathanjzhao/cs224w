```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get("https://www.usenix.org/conference/usenixsecurity24/call-for-papers")
soup = BeautifulSoup(response.text, parser="html.parser")
name = []
aff = []
role = []
year = []
sections = soup.find_all(attrs={"class": "grouping-field-heading"})
sections
```

```python
for section in sections:
  role_r = section.text.strip()
  row = section.next_sibling.next_sibling
  while not row is None and row["class"][0] != "grouping-field-heading":
    aff_r = row.find(attrs={"class": "field-content"}).find("em").text
    content = row.find(attrs={"class": "field-content"}).text.replace(u'\xa0', u' ')
    name_r = content.split(", ")[0]
    name.append(name_r)
    aff.append(aff_r)
    year.append("2024")
    role.append(role_r)
    row = row.next_sibling.next_sibling
```
