```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get("https://www.ndss-symposium.org/ndss2024/leadership/program-committee/")
soup = BeautifulSoup(response.text, parser="html.parser")
name = []
aff = []
role = []
year = []
sections = soup.find_all("h2", attrs={"class": "wp-block-heading"})
sections
```

```python
name.extend(["Mathias Payer", "Christina Pöpper"])
aff.extend(["EPFL", "New York University Abu Dhabi"])
role.extend(["PC Chair"] * 2)
year.extend(["2024"] * 2)
```

```python
rows = sections[1].next_sibling.next_sibling.find_all("p")
for row in rows:
  content= row.text.split(", ", 1)
  name.append(content[0])
  aff.append(content[1])
  role.append("PC Member")
  year.append("2024")
```

```python
response = requests.get("https://www.ndss-symposium.org/ndss2022/leadership/program-committee/")
soup = BeautifulSoup(response.text, parser="html.parser")
sections = soup.find_all("h2", attrs={"class": "wp-block-heading"})
sections
```

```python
name.extend(["Farinaz Koushanfar", "Wenyuan Xu"])
aff.extend(["University of California San Diego", "Zhejiang University"])
role.extend(["PC Chair"] * 2)
year.extend(["2022"] * 2)
```

```python
columns = sections[1].next_sibling.next_sibling.find_all("p")
for column in columns:
  rows = str(column).lstrip("<p>").rstrip("</p>").split("<br/>")
  for row in rows:
    content= row.split(", ", 1)
    name.append(content[0])
    aff.append(content[1])
    role.append("PC Member")
    year.append("2022")
```

```python
df = pd.DataFrame({"Name":name, "Affiliation":aff, "Role":role, "Year":year})
df
```

```python
df.to_csv("NDSS.csv", index=False)
```
